## Conditional Statements (`if`, `elif`, `else`)
Conditionals allow to choose between options or take different paths.

## Conditional Logic (if / else)

```mermaid
flowchart LR
    A[Start] --> B{Condition true?}
    B -- Yes --> C[Execute block A]
    B -- No --> D[Execute block B]
    C --> E[End]
    D --> E
```

## Multiple Conditions (if / else if / else)

```mermaid
flowchart LR
    A[Start] --> B{Condition 1?}
    B -- Yes --> C[Action 1]
    B -- No --> D{Condition 2?}
    D -- Yes --> E[Action 2]
    D -- No --> F[Default action]
    C --> G[End]
    E --> G
    F --> G
```


````python
temperature = 25

if temperature > 30:
    print("It's a hot day!")
elif temperature > 20:
    print("It's a nice day.")
else:
    print("It's a bit chilly.")
````

````python
x = 10

if x >= 50:
    print("A")
elif x >= 25:
    print("B")
else:
    print("C")
````

## For Loops
Use a `for` loop when you know how many times you want to run a block of code.

Note that while the underlying model is the same, Python `for` loops works a bit differently: they are controlled by iterators.

```python
for x in iterable:
    body
```

```mermaid
flowchart LR
    A[Start] --> B[Initialize counter]
    B --> C{Counter < limit?}
    C -- Yes --> D[Execute body]
    D --> E[Increment counter]
    E --> C
    C -- No --> F[End]
```



````python
lst = [1,2,3]

for i in lst:
    print(i)

list(range(5))
````

````python
print("Counting to 5:")
for i in range(1, 6):
    print(i)

print("\nIterating through a list:")
colors = ["Red", "Green", "Blue"]
for color in colors:
    print(f"Color: {color}")
````

## While Loops
Use a `while` loop when you want to repeat code as long as a condition is **True**.

```mermaid
flowchart LR
    A[Start] --> B{Condition true?}
    B -- Yes --> C[Execute loop body]
    C --> B
    B -- No --> D[Exit loop]
    D --> E[End]
```

````python
x = 3

while x > 1:
    print('wow')
    print(x)
    x = x -1
````

````python
energy = 3
while energy > 0:
    print(f"Running... Energy left: {energy}")
    energy -= 1
print("Stopped. Need to rest!")
````

## for and while usecases:

- `for`: Use when you want to repeat some action for a known number of times, while iterating over an iterator. Example: processing a **known** number of files in a folder.
- `while`: repeat some action for an unknown number of times, based on some condition.

## Break and Continue
* `break` exits the loop early.
* `continue` skips the current iteration and moves to the next.

````python
print("Testing Break (stops at 3):")
for n in range(1, 6):
    if n == 3:
        break
    print(n)

print("\nTesting Continue (skips 3):")
for n in range(1, 6):
    if n == 3:
        continue
    print(n)
````

````python
for n in range(6):
    # break    
    print(n)
    if n == 3:
        break
        # continue
````

````python
x = 'srm university-ap'
vowels ='aeiou'
# vowels = ['a', 'e', 'i', 'o', 'u']
for i in x:
    if i in vowels:
        print(i)
    continue
````
